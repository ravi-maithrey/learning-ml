```python
# importing csv data
from numpy import loadtxt
# import like the below two line for intellisense to work
from tensorflow import keras
import tensorflow as tf
# import like the above two lines for intellisense to work
from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset = loadtxt("../datasets/pima-indians-diabetes.data.csv", delimiter=',')
dataset.size
```

```python
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
#models in keras (in general as well) are a sequence of layers
model = Sequential()
model.add(Dense(12, input_shape=(8,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='tanh')) # added this to test it out
model.add(Dense(1, activation='sigmoid'))
```

```python
# compliling the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# now to fit the model
# training occurs over epochs, each epoch is split into batches
model.fit(X,y,epochs=150, batch_size=10)
```

```python
predictions = model.predict(X)
[round(x[0]) for x in predictions]
```
