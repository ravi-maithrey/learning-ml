```python
# import like the below two line for intellisense to work
from tensorflow import keras
import tensorflow as tf
# import like the above two lines for intellisense to work
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python
# we want to set aside some of our data automatically as validation set
dataset = np.loadtxt("../datasets/pima-indians-diabetes.data.csv", delimiter=",")
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
model = Sequential()
model.add(Dense(12,input_shape=(8,),activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='relu'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# note the validation_split
model.fit(X, y, validation_split=0.33, epochs=150, batch_size=10)
```
